```python
#list of michelin star'ed restaurants
msdf=pd.read_csv('michelin_star_restaurants2.csv')
msdf.head()
```

```python
#list of non-Michelin star restaurants
nodf=pd.read_csv('no_star_restaurants2.csv')
nodf.head()
```

```python
#adding stars to Michelin restaurants
result=pd.merge(msdf,star,how='left',on='id')
```

```python
result[result['star'].isnull()]
```

```python
#manally adding stars to those that are unassigned due to accented words on id
result['star'][43]=1
result['star'][49]=1
result['star'][92]=1
result['star'][96]=1
result['star'][105]=2
result['star'][122]=1
```

```python
print dcdf.shape
dcdf=dcdf.drop_duplicates()
dcdf=dcdf[dcdf.categories != 'categories']
print dcdf.shape
```

```python
trainer = result.append(nodf)
```

```python
trainer.shape
```

```python
#converting dollar signs to numbers
dcdf.ix[dcdf.price == '$$','price']=2
dcdf.ix[dcdf.price == '$$$','price']=3
dcdf.ix[dcdf.price == '$$$$','price']=4
```

```python
trainer.ix[trainer.price == '$$','price']=2
trainer.ix[trainer.price == '$$$','price']=3
trainer.ix[trainer.price == '$$$$','price']=4
```

```python
# from sklearn.preprocessing import LabelEncoder

# lbl = LabelEncoder()
# lbl.fit(list(dcdf['price'].values) + list(trainer['price'].values))
# dcdf['price'] = lbl.transform(dcdf['price'].values)
# trainer['price'] = lbl.transform(trainer['price'].values)
```

```python
#converting objects to numbers
dcdf['rating']=pd.to_numeric(dcdf['rating'])
dcdf['price']=pd.to_numeric(dcdf['price'])
dcdf['review_count']=pd.to_numeric(dcdf['review_count'])
dcdf['star']=pd.to_numeric(dcdf['star'])
trainer['rating']=pd.to_numeric(trainer['rating'])
trainer['price']=pd.to_numeric(trainer['price'])
trainer['review_count']=pd.to_numeric(trainer['review_count'])
trainer['star']=pd.to_numeric(trainer['star'])
dcdf.dtypes
```

```python
# from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
# ss=StandardScaler()
le=LabelEncoder()
# ss.fit(list(trainer['review_count'].values)+list(dcdf['review_count'].values))
# dcdf['review_count']=ss.transform(dcdf['review_count'].values)
# trainer['review_count']=ss.transform(trainer['review_count'].values)

le.fit(list(trainer['categories'].values)+list(dcdf['categories'].values))
dcdf['categories']=le.transform(dcdf['categories'].values)
trainer['categories']=le.transform(trainer['categories'].values)

# ss.fit(list(trainer['categories'].values)+list(dcdf['categories'].values))
# dcdf['categories']=ss.transform(dcdf['categories'].values)
# trainer['categories']=ss.transform(trainer['categories'].values)
```

```python
y_train=trainer[['star']]
X_train=trainer[['categories','price','rating','review_count']]
```

```python
y_test=dcdf[['star']]
X_test=dcdf[['categories','price','rating','review_count']]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier

#used random forest, extra tress, GBC, decision tree
svc = SVC().fit(X_train,y_train)
rf = RandomForestClassifier().fit(X_train,y_train)
gbc = GradientBoostingClassifier().fit(X_train, y_train)
dt = DecisionTreeClassifier(max_depth = 3, min_samples_split = 2).fit(X_train, y_train)
et = ExtraTreesClassifier(class_weight='balanced', n_jobs=-1).fit(X_train, y_train)

y_pred_svc=svc.predict(X_test)
y_pred_rf=rf.predict(X_test)
y_pred_gbc=gbc.predict(X_test)
y_pred_dt=dt.predict(X_test)
y_pred_et=et.predict(X_test)
y_pred_lr=et.predict(X_test)
```

```python
zip(dcdf['name'], y_pred_dt, y_pred_rf, y_pred_gbc, y_pred_et, y_pred_lr)
```

```python
#standardized rating, review_count, price based on mean and std for each city
trainer.ix[trainer.city=='New York','rating']= \
(trainer.ix[trainer.city=='New York','rating']-trainer.ix[trainer.city=='New York','rating'].mean())/ \
trainer.ix[trainer.city=='New York','rating'].std()

trainer.ix[trainer.city=='San Francisco','rating']= \
(trainer.ix[trainer.city=='San Francisco','rating']-trainer.ix[trainer.city=='San Francisco','rating'].mean())/ \
trainer.ix[trainer.city=='San Francisco','rating'].std()

trainer.ix[trainer.city=='Chicago','rating']= \
(trainer.ix[trainer.city=='Chicago','rating']-trainer.ix[trainer.city=='Chicago','rating'].mean())/ \
trainer.ix[trainer.city=='Chicago','rating'].std()

dcdf['rating']= (dcdf['rating']-dcdf['rating'].mean())/dcdf['rating'].std()

trainer.ix[trainer.city=='New York','review_count']= \
(trainer.ix[trainer.city=='New York','review_count']-trainer.ix[trainer.city=='New York','review_count'].mean())/ \
trainer.ix[trainer.city=='New York','review_count'].std()

trainer.ix[trainer.city=='San Francisco','review_count']= \
(trainer.ix[trainer.city=='San Francisco','review_count']-trainer.ix[trainer.city=='San Francisco','review_count'].mean())/ \
trainer.ix[trainer.city=='San Francisco','review_count'].std()

trainer.ix[trainer.city=='Chicago','review_count']= \
(trainer.ix[trainer.city=='Chicago','review_count']-trainer.ix[trainer.city=='Chicago','review_count'].mean())/ \
trainer.ix[trainer.city=='Chicago','review_count'].std()

dcdf['review_count']= (dcdf['review_count']-dcdf['review_count'].mean())/dcdf['review_count'].std()

trainer.ix[trainer.city=='New York','price']= \
(trainer.ix[trainer.city=='New York','price']-trainer.ix[trainer.city=='New York','price'].mean())/ \
trainer.ix[trainer.city=='New York','price'].std()

trainer.ix[trainer.city=='San Francisco','price']= \
(trainer.ix[trainer.city=='San Francisco','price']-trainer.ix[trainer.city=='San Francisco','price'].mean())/ \
trainer.ix[trainer.city=='San Francisco','price'].std()

trainer.ix[trainer.city=='Chicago','price']= \
(trainer.ix[trainer.city=='Chicago','price']-trainer.ix[trainer.city=='Chicago','price'].mean())/ \
trainer.ix[trainer.city=='Chicago','price'].std()

dcdf['price']= (dcdf['price']-dcdf['price'].mean())/dcdf['price'].std()
```

```python
#ran svc, random forest, gbc, decision tree, and extra trees
svc = SVC().fit(X_train,y_train)
rf = RandomForestClassifier().fit(X_train,y_train)
gbc = GradientBoostingClassifier().fit(X_train, y_train)
dt = DecisionTreeClassifier(max_depth = 3, min_samples_split = 2).fit(X_train, y_train)
et = ExtraTreesClassifier(class_weight='balanced', n_jobs=-1).fit(X_train, y_train)

y_pred_svc=svc.predict(X_test)
y_pred_rf=rf.predict(X_test)
y_pred_gbc=gbc.predict(X_test)
y_pred_dt=dt.predict(X_test)
y_pred_et=et.predict(X_test)

```
