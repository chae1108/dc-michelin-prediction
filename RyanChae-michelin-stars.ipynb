```python
#gets michelin-stared restaurants via Yelp api
import requests
import os
```

```python
def get_access_token():
    response = requests.post('https://api.yelp.com/oauth2/token',
                             data={'grand_type': 'client_credentials',
                             'client_id': 'XXXXX',
                             'client_secret': 'XXXXX'})
    return response.json()['access_token']
```

```python
def get_restaurants(rest_name):
    access_token = get_access_token()
    base_url = "https://api.yelp.com/v3/businesses/"
    rest_id = rest_name
    response = requests.get(base_url + str(rest_id),headers={'Authorization': 'Bearer %s' % access_token})
    restaurants = response.json()
    return restaurants
```

```python
import pandas as pd
yelpid=pd.read_csv('yelpid.csv')
yelpid
```

```python
def repeatlist(list_file):
    list_rest=[]
    for x in list_file:
        r_dict={}
        restaurant=get_restaurants(x)
        r_dict['name']=restaurant['name']
        r_dict['rating']=restaurant['rating']
        r_dict['price']=restaurant['price']
        r_dict['review_count']=restaurant['review_count']
        r_dict['categories']=restaurant['categories'][0]['alias']
        r_dict['id']=restaurant['id']
        r_dict['city']=restaurant['location']['city']
        r_dict['zip_code']=restaurant['location']['zip_code']
        list_rest.append(r_dict)
    return list_rest
```

```python
michelin_list=repeatlist(yelpid['Yelp'])
```

```python
michelin_list2=[michelin_list]
```
